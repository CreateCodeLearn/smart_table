### Function to get weights

```python
#%%writefile -a ./src/smarttable_data_processing.py

def get_weight(excel_filepath="./data/Sample Data.xlsx", subset=None, 
               relation_column="Relation degree", update_dict=None, *args, **kwargs):
    
    # helper functions
    def ones_dict(unique_relations):
        out = {}
        for item in unique_relations:
            out[item] = 1
        return out

    def edge_generator(d):    
        for item in d.keys():
            yield (item[0], 
                   item[1], 
                   dict(relation_weight=d.get(item, 1)))

    # load data
    data = pd.read_excel(excel_filepath, skiprows=2)
    # clean data
    data = data.applymap(lambda x: x.strip())
    data.columns = [x.strip() for x in data.columns]
    # Get relations
    relations = data[relation_column].unique()
    # unique relations
    from itertools import combinations
    unique_relations = list(combinations(relations, 2))
    relation_dict = ones_dict(unique_relations)
    # update relation based on user input
    if update_dict is not None:
        relation_dict.update(update_dict)
        
    # Create network
    import networkx as nx
    G = nx.Graph()
    G.add_edges_from(edge_generator(relation_dict))
    if subset is None:
        return G.size(weight='relation_weight')
    else:
        H = G.subgraph(subset)
        return H.size(weight='relation_weight')
        
```

```python
update_dict = {
    ('Bride', 'Groom'): 10,
    ('Bride', "Kim's Mom"): 7,
    ('Bride', "Kim's Dad"): 7,
    ('Bride', "Kim's Sister"): 3,
    ('Bride', "Kim's Brother"): 3,
    ('Bride', "Kim's Friend (work)"): 2,
    ('Bride', "Kanye's Mom"): 2,
    ('Bride', "Kanye's Dad"): 2,
    ('Bride', "Kanye's Friend (work)"): -2,
    ('Groom', "Kim's Mom"): 5,
    ('Groom', "Kim's Dad"): 5,
    ('Groom', "Kim's Friend (work)"): -2,
    ('Groom', "Kanye's Mom"): 7,
    ('Groom', "Kanye's Dad"): 7,
    ('Groom', "Kanye's Friend (work)"): 2,
}        
 
subset = ['Bride', 'Groom', "Kanye's Dad", "Kanye's Friend (work)", "Kim's Friend (work)"]
get_weight(subset=subset, update_dict=update_dict)
```

### Function to get relations based on names

```python
#%%writefile -a ./src/smarttable_data_processing.py
def names_2_relation_degree(names, 
                            excel_filepath="./data/Sample Data.xlsx", 
                            relation_column="Relation degree"):
    if isinstance(names, str):
        names = [names]
    # load data
    data = pd.read_excel(excel_filepath, skiprows=2)
    # clean data
    data = data.applymap(lambda x: x.strip())
    data.columns = [x.strip() for x in data.columns]
    relations = []
    for name in names:
        first = name.split()[0]
        second = name.split()[1]
        relation = data.loc[(data["First Name"] == first) & (data["Surname"]==second), relation_column]
        relations.append(relation.values[0])
    return relations    
```

```python
names_2_relation_degree(names=["Kim Kardashian", "Kris Jenner"])
```

### Function to evalute happiness for a subgroup of people

```python
#%%writefile -a ./src/smarttable_data_processing.py

def happiness(table, *args, **kwargs):
    """
    Find the happiness of the table
    """
    subset = names_2_relation_degree(table) 
    rv = get_weight(subset=subset, *args, **kwargs)
    return rv
```

```python
happiness(table=["Kim Kardashian", "Kris Jenner"], update_dict=update_dict)
```
